```python
import numpy as np
import pandas as pd
import os
import shutil
```

```python
folder_path = '/home/jupyter/gurih_mas/digit_recognizer/dataset/rotated/train'

images_train = pd.DataFrame([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
images_train.columns = ['filename']
images_train['image_flag'] = 'train'
images_train
```

```python
folder_path = '/home/jupyter/gurih_mas/digit_recognizer/dataset/rotated/test'

images_test = pd.DataFrame([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
images_test.columns = ['filename']
images_test['image_flag'] = 'test'
images_test
```

```python
images = pd.concat([images_train,images_test], axis=0)
images
```

```python
images.to_csv('/home/jupyter/gurih_mas/digit_recognizer/dataset/rotated.csv',index=False)
```
